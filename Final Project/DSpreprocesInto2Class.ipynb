```python
import os
import shutil

SRC = "/content/drive/My Drive/Simplified"     # your input folder
DST = "/content/drive/My Drive/dataset_skeleton"

correct_dir = os.path.join(DST, "correct")
incorrect_dir = os.path.join(DST, "incorrect")

os.makedirs(correct_dir, exist_ok=True)
os.makedirs(incorrect_dir, exist_ok=True)

for fname in os.listdir(SRC):
    if not fname.endswith(".txt"):
        continue

    parts = fname.split("_")  # example: 306_18_2_1_1_sit.txt

    if len(parts) < 6:
        print("Skipping unexpected filename:", fname)
        continue

    exercise_type = parts[2]          # 3rd index (0-based)
    correctness = parts[-2]           # second from last index

    # keep only files where exercise type == 2
    if exercise_type != "2":
        continue

    src_path = os.path.join(SRC, fname)

    if correctness == "1":    # correct
        dst_path = os.path.join(correct_dir, fname)
    elif correctness == "2":  # incorrect
        dst_path = os.path.join(incorrect_dir, fname)
    else:
        print("Skipping unknown correctness:", fname)
        continue

    shutil.copy(src_path, dst_path)

print("Done! Organized dataset into correct/incorrect.")
```
